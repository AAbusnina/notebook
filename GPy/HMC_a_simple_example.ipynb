```python
import numpy as np
import GPy
%matplotlib inline
#%config InlineBackend.figure_format = 'svg'
from pylab import *
```

# Example 1  HMC for inducing points with sparse GP

```python
m = GPy.examples.regression.sparse_GP_regression_1D()
m.likelihood.fix()
m.kern.fix()
```

### Run HMC

##### In order to run HMC in GP, we need to provide at least two parameters. stepsize is the numerical step size of solving hamilton equation. Every iteratin of HMC we will use hamilton equation to move the auxilary variable and target variable and make new proposal.  hmc_iters is number of moves we need for hamilton equation.

```python
m.Z.randomize()
# Sample the inducing inputs
hmc = GPy.inference.optimization.HMC(m, stepsize=1e-2)
t = hmc.sample(m_iters=1000,hmc_iters=20)
```

```python
print t[500:].mean(axis=0)
print t[500:].std(axis=0)
_=plot(t)
```

```python
_=scatter(t[500:,0],t[500:,1])
```

#### Run HMC_shortcut, we do not need to set stepsize for this version of HMC

```python
m.Z.randomize()
# Sample the inducing inputs
hmc = GPy.inference.optimization.HMC_shortcut(m, Hstd_th=[1e-8,3.])
t = hmc.sample(m_iters=1000,hmc_iters=20)
```

```python
print t[500:].mean(axis=0)
print t[500:].std(axis=0)
_=plot(t)
```

```python
_=scatter(t[500:,0],t[500:,1])
```

# Example 2 HMC for lengthscale and variance with marathon data

##### we set prior for lengthscale and variance of kernel. The mean of the prior is close to the result of GP optimisation. we then allow a big variance. In the case below,we set gamma prior to lengthscale and variance. E(lengthscale) = 120, Var(lengthscale)=2000, E(variance) = 25, Var(variance) = 150 

```python
m = GPy.examples.regression.olympic_marathon_men()
#
#set prior for lengthscale and variance.
m.kern.variance.set_prior(GPy.core.parameterization.priors.Gamma.from_EV(25.,150.))
m.kern.lengthscale.set_prior(GPy.core.parameterization.priors.Gamma.from_EV(120.,2000.))
m.optimize('scg',messages=1,max_iters=10000)
print m
#_ = m.plot(plot_limits=(1850, 2050))
```

### Run HMC

##### we plot the full length of hmc iteration. The first 5000 could be burn in stage and can be ignored. 

```python
m.randomize()
# initialise hmc
hmc = GPy.inference.optimization.HMC(m,stepsize=2e-1)
# run hmc
t = hmc.sample(m_iters=20000,hmc_iters=20)

# Sample parameters 
#hmc = GPy.inference.optimization.HMC(m, stepsize=5e-1)
#t = hmc.sample(m_iters=50000,hmc_iters=20)

_=plot(t)
```

```python
_=plot(t[:,0],t[:,1],'.')
```
